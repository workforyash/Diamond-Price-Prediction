```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
x=0
for i in numTypes:
   sns.histplot(data=df, x=i, kde=True)
   print('\n')
   plt.show() 
```

```python
sns.heatmap(df[numTypes].corr(), annot=True)
```

```python
## mapping the categorical values according to the rank given by the american gem society
df['cut'].unique()

cut_map = {
    'Premium':4, 
    'Very Good':2, 
    'Ideal':5, 
    'Good':1, 
    'Fair':3
}
```

```python
df['color'].unique()

color_map={
    'F':3, 
    'J':7, 
    'G':4, 
    'E':2, 
    'D':1, 
    'H':5, 
    'I':6
}
```

```python
df['clarity'].unique()

clarity_map = {
    'VS2':5, 
    'SI2':3, 
    'VS1':6, 
    'SI1':4, 
    'IF':2, 
    'VVS2':7, 
    'VVS1':8, 
    'I1':1
}
```

```python
df['cut'] = df['cut'].map(cut_map)
df['color'] = df['color'].map(color_map)
df['clarity'] = df['clarity'].map(clarity_map)
```
